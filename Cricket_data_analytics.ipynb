```python
#importing libraries
import pandas as pd
import json
```

```python
with open("/content/sample_data/sports/t20_wc_match_results.json") as f:
  data=json.load(f)
```

```python
data
```

```python
df_match=pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard':'match_id'},inplace=True,axis=1)
```

```python
df_match.head()
```

```python
match_id_dict={}
for index, row in df_match.iterrows():
  key1=row['team1'] + ' Vs ' + row['team2']
  key2=row['team2'] + ' Vs ' + row['team1']
  match_id_dict[key1]=row['match_id']
  match_id_dict[key2]=row['match_id']

match_id_dict
```

```python
df_match.to_csv('/content/sample_data/sports/df_match_summary.csv', index = False)
```

Batting summary

```python
with open("/content/sample_data/sports/t20_wc_batting_summary.json") as f:
  data_batting=json.load(f)
  all_records=[]
  for rec in data_batting:
    all_records.extend(rec['battingSummary'])
```

```python
data_batting=pd.DataFrame(all_records)
data_batting.head()
```

```python
data_batting['out/not_out']=data_batting.dismissal.apply(lambda x:"out" if len(x)>0 else "not_out")
data_batting.head(10)
```

```python
data_batting.drop(columns=['dismissal'],inplace=True)
data_batting.head(10)
```

```python
data_batting['batsmanName']=data_batting['batsmanName'].apply(lambda x:x.replace('',''))
data_batting['batsmanName']=data_batting['batsmanName'].apply(lambda x:x.replace('\xa0',''))
data_batting.head(20)
```

```python
data_batting['match_id']=data_batting['match'].map(match_id_dict)
data_batting.head()
```

```python
data_batting.to_csv('/content/sample_data/sports/fact_batting_summary.csv', index = False)
```
